```python
locations = list(set(df['location']))
new_df = pd.DataFrame(index = df.columns[:-1], columns = locations)

import numpy as np
l = [[] for i in range(4)]
for i in new_df.index:
    for j in locations:
        new_df.loc[i][j] = [].copy()
```

```python
for i in df.index:
    for j in new_df.index:
        # print(i,j)
        new_df.loc[j][df.iloc[i]['location']].append(df.iloc[i][j])
```

```python
# import SentimentIntensityAnalyzer class from vaderSentiment.vaderSentiment module.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#vader is good for social media data analysis
# function to print sentiments of the sentence.
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(sentence)
    return  sentiment_dict['neg'];
```

```python

for i in new_df.index: 
    for j in new_df.columns: 
        # print(i,",",j,",",new_df.loc[i][j])
        q = new_df.loc[i][j]
        p = []
        for k in q: 
            p.append(sentiment_scores(k))
        # print(i,",",j)    
        # print(p)
        
        avg=np.average(p)
        q.append(avg)
        # print(q)
        new_df.loc[i][j]=avg


        
```

```python
  
livability_df = pd.DataFrame(index=new_df.columns, columns=['average of negativity', 'average of positivity'])
for i in new_df.columns:
    q=[]
    for j in new_df.index:
        q.append(new_df.loc[j][i])
      
    avg_liv=np.average(q)
    livability_df.loc[i]['average of negativity'] = ((avg_liv))
    livability_df.loc[i]['average of positivity'] = (1-(avg_liv))
livability_df 
```

```python
percent_df = pd.DataFrame(index=new_df.index, columns=new_df.columns)
for i in new_df.index:
    q=[]
    for j in new_df.columns:
        q.append(new_df.loc[i][j])
    summ = sum(q)
    #print(summ)
    #percentlist=[]
    for j in new_df.columns:
        percent = (new_df.loc[i][j]/summ)*100
        percent_df.loc[i][j] = percent
        #percentlist.append(percent)
    #print(sum(percentlist))
```

```python
import matplotlib.pyplot as plt

# Define a color palette with 4 colors
colors = ['#3E5C76', '#548CA8', '#8AB5C9', '#C9E0E1']

# Create a 4 by 4 grid of subplots with a larger size
fig, axs = plt.subplots(4, 4, figsize=(30, 30))

# Flatten the axs array for easy iteration
axs = axs.flatten()

# Iterate over the columns of the new_df DataFrame and create a pie chart for each column
for i, (column, series) in enumerate(new_df.iteritems()):
    # Plot the pie chart on the current axis, with a larger radius
    series.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colors, ax=axs[i], radius=1.5)
    axs[i].set_title(column)
    axs[i].set_ylabel('')
    axs[i].set_xlabel('')
    
# Adjust the layout of the subplots and show the plot
plt.tight_layout()
plt.show()
```
